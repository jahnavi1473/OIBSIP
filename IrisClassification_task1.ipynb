**IRIS FLOWER CLASSIFICATION**

Iris flower has three species-setosa,versicolor and virginica based on
different measurements of sepals and petal width.

------------------------
The aim is to train a machine learning model that learns from the measurements and classifies them among the 3 species.

------------------------
Data source is iris.csv which contains 5 columns and 150 rows.

IMPORTING LIBRARIES


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

LOADING THE DATASET

```python
iris_data=pd.read_csv('Iris.csv')
iris_data.head(5)
```

```python
iris_data.describe()
```

```python
iris_data=iris_data.drop("Id",axis=1) # removing column "id" since it is not required
```

```python
iris_data.shape
```


*   contains 5 columns and 150 rows.
*   "id" column is removed






```python
iris_data.head(5)
```

```python
iris_data.isnull().sum()
```

no null values so we can directly split the data

```python
corr=iris_data.corr()
print(corr)
```

```python
fig,ax=plt.subplots(figsize=(5,5))
sns.heatmap(corr,annot=True,ax=ax)
```

separating input columns and target value

```python
x = iris_data.iloc[:, :-1]
y = iris_data.iloc[:,-1]
```

```python
x
```

```python
y
```

Splitting the data into training and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

Model - logistic regression

```python
from sklearn.linear_model import LogisticRegression
lrmodel=LogisticRegression()
lrmodel.fit(x_train,y_train)
```

```python
y_pred=lrmodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print("Confusion matrix is:\n",confusion_matrix(y_test, y_pred),"\n")
print("Accuracy is:",accuracy_score(y_test,y_pred)*100,"\n")
print("Classification report is:\n",classification_report(y_test,y_pred))
```

```python
newdata=np.array([[2.6,3,1,2.3],[4,2.5,3,2.9],[1.3,0.6,1.4,0.7]])
pred=lrmodel.predict(newdata)
print("Prediction of species is:{}".format(pred))
```
